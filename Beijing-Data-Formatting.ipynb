```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data.csv")
```

```python
data['hour_formatted'] = data.apply(lambda x: str(x['hour']) + ':00:00', axis=1)
```

```python
data['timestamp'] = data.apply(lambda x: str(x['year']) + '-' + str(x['month']) + '-' + str(x['day']) + ' ' + x['hour_formatted'], axis=1)
```

```python
data['timestamp'] = data.apply(lambda x: pd.to_datetime(x.timestamp), axis=1)
```

```python
data = data.drop(['hour_formatted', 'year', 'month', 'day', 'hour'], axis=1)
```
